```python
#group data by plot_id and by sex
by_site_sex = surveys_df.groupby(['plot_id', 'sex'])
```

```python
#show what this looks like, but ensure showing numeric only 
by_site_sex.mean(numeric_only=True)
```

```python
#now we just want the weight column and we want to show the total weight for each sex in each site
#give the object a name,call the previous object, limit this to the relevant column, define how this column will be aggregated

site_sex_count = by_site_sex['weight'].sum()
site_sex_count
```

```python
spc = site_sex_count.unstack()
spc
```

```python
#draw a bar chart which is stacked
#add axes labels and a title

s_plot = spc.plot(kind='bar', stacked=True)
s_plot.set_ylabel("Weight")
s_plot.set_xlabel("Plot")
s_plot.set_title("Total weight by plot and sex")
```
