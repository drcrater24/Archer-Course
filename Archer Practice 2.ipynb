```python
import pandas as pd
surveys_df = pd.read_csv('surveys.csv')
surveys_df
```

```python
surveys_df.columns
```

```python
site_names = pd.unique(surveys_df['plot_id'])
print(site_names)
len(site_names)
```

```python
site_names2 = surveys_df['plot_id'].nunique()
print(site_names2)
```

```python
surveys_df.describe()
```

```python
grouped_data = surveys_df.groupby('sex')
grouped_data.mean(numeric_only=True)
```

```python
grouped_data = surveys_df.groupby('sex')
grouped_data.describe()
```

```python
grouped_data2 = surveys_df.groupby(['plot_id','sex'])
grouped_data2.mean(numeric_only=True)
```

```python
by_site = surveys_df.groupby('plot_id')
by_site['weight'].describe()
```

```python
# Count the number of samples by species
species_counts = surveys_df.groupby('species_id')['record_id'].count()
print(species_counts)
```

```python
# Make sure figures appear inline in Ipython Notebook
%matplotlib inline
# Create a quick bar chart
species_counts.plot(kind='bar');
```

```python
total_count = surveys_df.groupby('plot_id')['record_id'].nunique()
# Let's plot that too
total_count.plot(kind='bar');
```

```python
species_weight = surveys_df.groupby('plot_id').agg({"weight": 'mean'})
print(species_weight)
```

```python
species_weight.plot(kind='bar');
```
